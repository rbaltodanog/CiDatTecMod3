1. ¿Cuál es la diferencia entre una función de densidad de probabilidad y una función de distribución de probabilidad acumulativa? Utilice un gráfico ilustrativo para la explicación. ¿Cuándo queremos usar una o la otra? ¿En el caso discreto, cuál es la diferencia en la interpretación del valor retornado por cada una de las dos funciones para la misma entrada $x_i$? **10 puntos**

La función de distribución de probabilidad acumulativa (CDF) expresa un rango de que un valor $X$ tome un valor menor o igual a un número real $x$ (es una integral como se dijo en clase y se acumulan los valores de probabilidades hasta el punto x). La diferencia entre la función de densidad de probabilidad (PDF) radica en lo que expresan:
- La CDF expresa acumulación: Probabilidad de $-\infty$ a x
- La PDF brinda la densidad de probabilidad que al integrarla sobre un intervalo se obtiene una probabilidad

Se puede observar en el siguiente gráfico de la clase 1:

![](https://www.vosesoftware.com/riskwiki/images/image15_400.gif)

La línea azul que es CDF (F(x)) acumula probabilidades y sigue un patrón con respecto a la línea roja que es PDF (f(x)) → la curva es gaussiana y en punto mas alto deja de acumular, haciendo que la curva se aplane para F(x).

Cuándo usar una o la otra:
- Usamos la PDF cuando queremos conocer qué tan probable es un valor específico o ver la forma de la distribución (dónde se concentran los datos, si hay asimetría, etc.).
- Usamos la CDF cuando queremos calcular probabilidades acumuladas, por ejemplo: $P(X \leq x)$, $P(X > x) = 1 - F(x)$, o $P(a \leq X \leq b) = F(b) - F(a)$.

En el caso discreto, para una misma entrada $x_i$:

La PMF (función de masa de probabilidad, análoga a la PDF) retorna $P(X = x_i)$, es decir, la probabilidad exacta de que la variable tome ese valor específico.
La CDF retorna $P(X \leq x_i)$, es decir, la probabilidad acumulada de que la variable tome un valor menor o igual a $x_i$ (la suma de todas las probabilidades desde el menor valor posible hasta $x_i$).

- Hipótesis nula ($H_0$): los datos provienen de la distribución especificada.

La funcion acumulativa ideal sería una estimación de parámetros ideales de la distribución propuesta: normal ($\sigma$, $\mu$), gamma ($\alpha$, $\beta$), u otras.

Luego de esto se calcula la distancia máxima entre las 2 functiones acumulativas y se verifica si esta excede el punto crítico dado por $\alpha$:
\begin{equation}
D=\max\left|F_e(x)-F_i(x)\right|
\end{equation}

6. Bajo el contexto de prueba de hipótesis, explique en que consiste cada tipo de error. Adicionalmente, explique que representan $\alpha$ y $\beta$ y como pueden reducirse. **10 puntos**

En pruebas de hipótesis, al usar estimaciones muestrales para aceptar o rechazar la hipótesis nula $H_0$, es posible cometer dos tipos de error:

**Tipos de error:**

| Decisión/Valor real | $H_0$ es verdadera | $H_0$ es falsa |
|---|---|---|
| Se acepta $H_0$ | Verdadero negativo (sin error) | **Falso negativo (error tipo II)** |
| Se rechaza $H_0$ | **Falso positivo (error tipo I)** | Verdadero positivo (sin error) |

**Error Tipo I ($\alpha$):**

Es la probabilidad de rechazar $H_0$ cuando $H_0$ es verdadera. Se conoce como el **nivel de significancia** o **error-$\alpha$**:

$$\alpha=Pr(\text{rechazar } H_0 \text{ cuando } H_0 \text{ es verdadera})$$

**Error Tipo II ($\beta$):**

Es la probabilidad de aceptar $H_0$ cuando $H_0$ es falsa:

$$\beta=Pr(\text{aceptar } H_0 \text{ cuando } H_0 \text{ es falsa})$$

**Cómo reducir $\alpha$ y $\beta$:**

1. **Ampliar la región de aceptación (RA)**: Reduce $\alpha$ pero aumenta $\beta$

2. **Incrementar el tamaño de la muestra $n$**: Reduce tanto $\alpha$ como $\beta$ simultáneamente (manteniendo constantes los valores críticos)

9. ¿Qué quiere decir que dos variables aleatorias son independientes y cómo se expresa matemáticamente? Dos variables aleatorias son independientes si y solo si la probabilidad conjunta de que $X=x$ y $Y=y$ se puede expresar como:

>>$p\left(X,Y\right)=p\left(X\right)p\left(Y\right).$

Y por regla del producto se puede expresar como 

>>$p\left(Y\right)=p\left(Y|X\right).$

10. Si se tienen 2 eventos $C$ y $D$ con probabilidades $p_C$ y $p_D$ ¿Cuál es la probabilidad máxima de que no ocurra $D$ o que ocurra $C$, en términos de sus probabilidades individuales? **5 puntos**

Buscamos la probabilidad máxima de $P(D^c \cup C)$.

Por la regla de la suma:

>>$P(\text{No Ocurra D} \cup C) = P(\text{No ocurra D}) + P(C) - P(\text{No ocurra D} \cap C)$

Sabemos que $P(\text{No ocurra D}) = 1 - p_D$, entonces:

>>$P(\text{No Ocurra D} \cup C) = (1 - p_D) + p_C - P(\text{No ocurra D} \cap C)$

Para **maximizar** esta expresión, debemos **minimizar** $P(\text{No ocurra D} \cap C)$. El valor mínimo de una probabilidad es $0$, lo cual ocurre cuando $C \subseteq D$ (es decir, todo evento donde ocurre $C$ también ocurre $D$, por lo que $C$ y $\text{No ocurra D}$ no se intersectan).

Por lo tanto, la probabilidad máxima es:

>>$P_{\max}(\text{No ocurra D} \cup C) = 1 - p_D + p_C$

Esto aplica siempre que $p_C \leq p_D$. Si $p_C > p_D$, la probabilidad máxima simplemente es $1$.

En general:

>>$P_{\max}(\text{No ocurra D} \cup C) = \min(1,\; 1 - p_D + p_C)$

11. ¿Cuáles fueron las medidas de tendencia central vistas en clase? ¿En qué consiste cada una? y ¿En qué casos es útil usar cada una? **10 puntos**

Las medidas de tendencia central fueron las siguientes (fuente: Lección 4):

| Distribución | Tipo | Media ($\mu$) | Varianza ($\sigma^2$) |
|---|---|---|---|
| Uniforme discreta | Discreta | $\dfrac{n+1}{2}$ | $\dfrac{n^2-1}{12}$ |
| Uniforme continua | Continua | $\dfrac{a+b}{2}$ | $\dfrac{(b-a)^2}{12}$ |

|---|---|

| **Uniforme continua** | Cualquier valor en un intervalo $[a,b]$ es igualmente probable |

12. Explique: ¿Qué es $SS_{T}$ en el contexto de ANOVA? ¿Cuáles son sus componentes y qué significado tiene cada uno? ¿Qué son y que relación tiene $\textrm{MS}_\textrm{tratamiento}$ y $\textrm{MS}_\textrm{error}$ con la hipótesis nula planteada en el ANOVA? **10 puntos**

**¿Qué es $SS_T$?**

$SS_T$ es la **suma total de cuadrados** que mide la variación total en los datos. Se descompone en dos componentes:

$$SS_T = SS_{\text{tratamiento}} + SS_{\text{error}}$$

**Componentes:**

1. **$SS_{\text{tratamiento}}$**: Mide la variación **entre grupos** (diferencias entre las medias de cada tratamiento y la media global). Representa cuánto de la variación se explica por las diferencias entre tratamientos.

2. **$SS_{\text{error}}$**: Mide la variación **dentro de cada grupo** (diferencias entre observaciones individuales y su media de grupo). Representa la variación aleatoria o no explicada.

**Relación con la hipótesis nula:**

- $\textrm{MS}_{\text{tratamiento}} = \frac{SS_{\text{tratamiento}}}{a-1}$ (media cuadrática del tratamiento)
- $\textrm{MS}_{\text{error}} = \frac{SS_{\text{error}}}{a(n-1)}$ (media cuadrática del error)

El estadístico $F = \frac{\textrm{MS}_{\text{tratamiento}}}{\textrm{MS}_{\text{error}}}$ compara estas dos fuentes de variación:

- Si $H_0$ es cierta (todas las medias son iguales), ambos estiman $\sigma^2$ y $F \approx 1$
- Si $H_0$ es falsa (al menos una media difiere), $\textrm{MS}_{\text{tratamiento}}$ será mayor y $F$ será significativamente grande, llevando al rechazo de $H_0$

13. Suponga que tenemos dos variables aleatorias $X_1$ y $X_2$ que representan la cantidad de vallas publicitarias compradas en la autopista Florencio de Castillo y el volumen de ventas por semana, respectivamente. Calcule la matriz de covarianza y la matriz de correlación de Pearson para una muestra a lo largo de 10 semanas con los siguientes datos: **20 puntos**

  - $h_1 = [2, 5, 1, 3, 4, 1, 5, 3, 4, 2]$
  - $h_2 = [50, 57, 41, 54, 54, 38, 63, 48, 59, 46] $

Para la realización de la matrix de covarianza determinada se procede a definir los vectores de longitud al igual que las variables aleatorias:

\begin{equation}
U=\left\{ \vec{h}_{1},\vec{h}_{2} \right\} =\begin{bmatrix}| & |\\
\vec{h}_{1} & \vec{h}_{2}\\
| & |
\end{bmatrix}=\begin{bmatrix}2 & 5 & 1 & 3 & 4 & 1 & 5 & 3 & 4 & 2\\
50 & 57 & 41 & 54 & 54 & 38 & 63 & 48 & 59 & 46
\end{bmatrix}
\end{equation}

con $\vec{u}_{i}\in\mathbb{R}^{2}$, donde cada dimensión es una variable aleatoria, y $U \in\mathbb{R}^{2\times10}$.

Se procede entonces a calcular las entradas $\Sigma_{H_{1},H_{2}}$, además de los valores de la diagonal $\sigma_{H_{1}}^{2}$ y $\sigma_{H_{2}}^{2}$ teniendo en cuenta que $\overline{\mu}_{H_{1}}=3$ y $\overline{\mu}_{H_{2}}=51$:

\begin{equation}
\begin{array}{c}
\Sigma_{H_{1},H_{2}}=\frac{1}{10-1}\left(\left(3-2\right)\left(51-50\right)+\left(3-5\right)\left(51-57\right)+\left(3-1\right)\left(51-41\right)+\left(3-3\right)\left(51-54\right)+\left(3-4\right)\left(51-54\right)+\left(3-1\right)\left(51-38\right)\left(3-5\right)\left(51-63\right)\left(3-3\right)\left(51-48\right)\left(3-4\right)\left(51-59\right)\left(3-2\right)\left(51-46\right)\right)\\

\sigma_{H_{1}}^{2}=\frac{1}{10-1}\left(\left(3-2\right)^{2}+\left(3-5\right)^{2}+\left(3-1\right)^{2}+\left(3-3\right)^{2}+\left(3-4\right)^{2}+\left(3-1\right)^{2}+\left(3-5\right)^{2}+\left(3-3\right)^{2}+\left(3-4\right)^{2}+\left(3-2\right)^{2}\right)\\

\sigma_{H_{2}}^{2}=\frac{1}{10-1}\left(\left(51-50\right)^{2}+\left(51-57\right)^{2}+\left(51-41\right)^{2}+\left(51-54\right)^{2}+\left(51-54\right)^{2}+\left(51-38\right)^{2}+\left(51-63\right)^{2}+\left(51-48\right)^{2}+\left(51-59\right)^{2}+\left(51-46\right)^{2}\right)\\
\end{array}
\end{equation}

lo cual desarrollado corresponde a:

\begin{equation}
\begin{array}{c}
\Sigma_{H_{1},H_{2}}=\frac{1}{9}\left((1)(1)+(-2)(-6)+(2)(10)+(0)(-3)+(-1)(-3)+(2)(13)+(-2)(-12)+(0)(3)+(-1)(-8)+(1)(5)\right)=\frac{99}{9}=11\\
\sigma_{H_{1}}^{2}=\frac{1}{9}\left(1+4+4+0+1+4+4+0+1+1\right)=\frac{20}{9}\approx2.2222\\
\sigma_{H_{2}}^{2}=\frac{1}{9}\left(1+36+100+9+9+169+144+9+64+25\right)=\frac{566}{9}\approx62.8888\\
\end{array}
\end{equation}

Por lo que se obtiene la matriz de covarianza:

\begin{equation}
\Sigma=\begin{bmatrix}\frac{20}{9} & \frac{99}{9}\\
\frac{99}{9} & \frac{566}{9} \\
\end{bmatrix}=\begin{bmatrix}2.2222 & 11\\
11 & 62.8888 \\
\end{bmatrix}.
\end{equation}

Para calcular la matriz de correlación de Pearson, se normaliza la covarianza dividiendo por el producto de las desviaciones estándar:

\begin{equation}
\rho_{H_{1},H_{2}}=\frac{\textrm{cov}\left(H_{1},H_{2}\right)}{\sigma_{H_{1}}\sigma_{H_{2}}}
\end{equation}

Calculando las desviaciones estándar:

\begin{equation}
\sigma_{H_{1}}=\sqrt{\frac{20}{9}}\approx1.4907
\end{equation}

\begin{equation}
\sigma_{H_{2}}=\sqrt{\frac{566}{9}}\approx7.9304
\end{equation}

Calculando el coeficiente de correlación de Pearson:

\begin{equation}
\rho_{H_{1},H_{2}}=\frac{11}{1.4907\times7.9304}=\frac{11}{11.8205}\approx0.9306
\end{equation}

Por lo que se obtiene la matriz de correlación de Pearson:

\begin{equation}
\rho=\begin{bmatrix}1 & 0.9306\\
0.9306 & 1
\end{bmatrix}
\end{equation}

Este coeficiente de 0.9306 indica una correlación positiva muy fuerte entre la cantidad de vallas publicitarias compradas y el volumen de ventas por semana.
