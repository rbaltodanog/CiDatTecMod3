## Parte 1

Cuatro plantas químicas, que producen los mismos productos y son propiedad de la misma compañía,
descargan aguas residuales en arroyos en la cercanía donde se encuentran ubicadas. Para vigilar la magnitud de la contaminación creada por las aguas residuales y para determinar si esto difiere de una planta a otra, la compañía recolectó aleatoriamente  cinco muestras (observaciones) de desechos líquidos de cada planta. Cada observación tomada indica la cantidad de libras por galón de desechos. Los datos se presentan en la siguiente tabla:

| Planta | Ob. 1 | Ob. 2 | Ob. 3 | Ob. 4 | Ob. 5 |
|--------|-------|-------|-------|-------|-------|
| **A**  | 1.65  | 1.72  | 1.50  | 1.37  | 1.60  |
| **B**  | 1.70  | 1.85  | 1.46  | 1.46  | 1.80  |
| **C**  | 1.40  | 1.75  | 1.38  | 1.38  | 1.55  |
| **D**  | 2.10  | 1.95  | 1.65  | 1.65  | 1.88  |

¿La información aporta suficiente evidencia para indicar una diferencia en el peso medio de las aguas residuales por galón descargadas de las cuatro plantas? Pruebe usando $\alpha =0.05$ (70 puntos)

## Descripción del experimento

Se mide: **cantidad de libras por galón de desechos**

- **Variable nominal:** Planta de tratamiento  
- **Variable métrica:** Cantidad de lbs por galón de desecho

## Cálculo de la media total

$$
\mu = \frac{
1.65 + 1.72 + 1.50 + 1.37 + 1.60 +
1.70 + 1.85 + 1.46 + 1.46 + 1.80 +
1.40 + 1.75 + 1.38 + 1.38 + 1.55 +
2.10 + 1.95 + 1.65 + 1.65 + 1.88
}{20}
$$

$$
\mu = \frac{32.8}{20} = 1.64
$$

---

## Media de las observaciones por planta

### Planta A
$$
\bar y_A = \frac{1.65 + 1.72 + 1.50 + 1.37 + 1.60}{5}
= \frac{7.84}{5} = 1.568
$$

### Planta B
$$
\bar y_B = \frac{1.70 + 1.85 + 1.46 + 1.46 + 1.80}{5}
= \frac{8.27}{5} = 1.654
$$

### Planta C
$$
\bar y_C = \frac{1.40 + 1.75 + 1.38 + 1.38 + 1.55}{5}
= \frac{7.46}{5} = 1.492
$$

### Planta D
$$
\bar y_D = \frac{2.10 + 1.95 + 1.65 + 1.65 + 1.88}{5}
= \frac{9.23}{5} = 1.846
$$

---

## ANOVA de un factor

### Estadístico F

$$
F_0 = \frac{MS_{\text{tratamiento}}}{MS_{\text{error}}}
$$

---

## Suma total de cuadrados

$$
SS_T =
\sum_{i=1}^{a} \sum_{j=1}^{n} (y_{ij} - \bar y_i)^2
+ n \sum_{i=1}^{a} (\bar y_i - \bar y_{..})^2
$$

---

## Suma de cuadrados del tratamiento

$$
SS_{\text{tratamiento}} =
n \sum_{i=1}^{a} (\bar y_i - \bar y_{..})^2
$$

$$
= 5 \Big[
(1.568 - 1.64)^2 +
(1.654 - 1.64)^2 +
(1.492 - 1.64)^2 +
(1.846 - 1.64)^2
\Big]
$$

$$
= 5 (0.005184 + 0.000196 + 0.021904 + 0.042436)
$$

$$
SS_{\text{tratamiento}} = 0.3485
$$

---

## Media cuadrática del tratamiento

$$
MS_{\text{tratamiento}} =
\frac{SS_{\text{tratamiento}}}{a - 1}
=
\frac{0.3485}{4 - 1}
=
0.1162
$$

---

## Suma de cuadrados del error

$$
SS_{\text{error}} =
\sum_{i=1}^{a} \sum_{j=1}^{n} (y_{ij} - \bar y_i)^2
$$

$$
\begin{aligned}
SS_{\text{error}} ={}&
(1.65-1.568)^2 + (1.72-1.568)^2 + (1.50-1.568)^2 + (1.37-1.568)^2 + (1.60-1.568)^2 \\
&+ (1.70-1.654)^2 + (1.85-1.654)^2 + (1.46-1.654)^2 + (1.46-1.654)^2 + (1.80-1.654)^2 \\
&+ (1.40-1.492)^2 + (1.75-1.492)^2 + (1.38-1.492)^2 + (1.38-1.492)^2 + (1.55-1.492)^2 \\
&+ (2.10-1.846)^2 + (1.95-1.846)^2 + (1.65-1.846)^2 + (1.65-1.846)^2 + (1.88-1.846)^2
\end{aligned}
$$

$$
SS_{\text{error}} = 0.4686
$$

---

## Media cuadrática del error

$$
MS_{\text{error}} =
\frac{SS_{\text{error}}}{a(n-1)}
=
\frac{0.4686}{4(5-1)}
=
0.0292875
$$

---

## Estadístico F observado

$$
F_0 =
\frac{0.1162}{0.0292875}
=
3.97
$$

---

## Decisión estadística

Con:
- $\alpha = 0.05$
- $a = 4$
- $N = 20$

$$
F_{\text{crítico}} \approx 3.24
$$

$$
F_0 = 0.3967 < 3.24
$$

**Se rechaza la hipótesis nula.**

Existe evidencia estadísticamente significativa para indicar que hay diferencias en el peso medio de las aguas residuales por galón descargadas de las cuatro plantas.
