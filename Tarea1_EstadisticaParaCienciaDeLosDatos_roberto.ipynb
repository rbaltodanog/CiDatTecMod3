1. Muestreo de estaturas. Los estudiantes deberán programar dos funciones para crear muestras de $N$ personas (observaciones) con una sola variable aleatoria: estatura. Para efectos prácticos, ambas funciones pueden retornar una simple lista con valores numéricos que representan la estatura en centímetros de diferentes personas. Para ambas funciones deberán mostrarse ejemplos de creación de muestras con al menos 100 observaciones y mostrarlas en un histograma. La cantidad de cubetas deberá limitarse a un máximo de 9. Las funciones a programar serán las siguientes:
 * **Muestreo Uniforme**. Además de recibir la cantidad de observaciones a generar $N$, recibirá la estatura mínima y máxima  por parámetro. Por ejemplo, si se pide una distribución uniforme entre 150 y 168, las estaturas entre ese rango deberán ser igual de probables de muestrear. (5 puntos)
  * **Muestreo Normal**. Además de recibir la cantidad de observaciones a generar $N$, recibirá la media y desviación estándar por parámetro. De esta forma, la función generará $N$ observaciones provenientes de una distribución Gaussiana con la media y desviación estándar recibidas. (5 puntos)

```python
import numpy as np
import matplotlib.pyplot as plt

def muestreo_uniforme(N, minima, maxima):
    rng = np.random.default_rng()
    return rng.uniform(minima, maxima, N)

def muestreo_normal(N, mu, sigma):
    rng = np.random.default_rng()
    return rng.normal(mu, sigma, N)

# Parametros de las funciones
minima = 150
maxima = 168
mu = 170
sigma = 5
N = 100

# Muestreo uniforme entre minima y maxima
estaturas_uniformes = muestreo_uniforme(N, minima, maxima)

# Muestreo normal con media y desviación
estaturas_normales = muestreo_normal(N, mu, sigma)

# Crear histogramas con máximo 9 cubetas
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(estaturas_uniformes, bins=9, alpha=0.7, color='blue', edgecolor='black')
plt.title(f'Distribución Uniforme ({minima}-{maxima} cm)')
plt.xlabel('Estatura (cm)')
plt.ylabel('Frecuencia')

plt.subplot(1, 2, 2)
plt.hist(estaturas_normales, bins=9, alpha=0.7, color='red', edgecolor='black')
plt.title(f'Distribución Normal (mu={mu}, sigma={sigma})')
plt.xlabel('Estatura (cm)')
plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

2. **Evaluación de la función de verosimilitud normal**. Deberán proveer una función que reciba una de las **muestras** del punto 1 y evalúe la probabilidad que esa muestra provenga de una distribución normal con parámetros $\mu$ y $\sigma$ (que deberán ser enviados a la función). Deberán mostrar ejemplos obtenidos bajo ambos esquemas de muestreo. Por ejemplo, si se crea una muestra normal centrada en $\mu_1=177$ con $\sigma_1=1$ y evaluamos la verosimilitud para $\mu=176,\sigma=3$ podría darnos una probabilidad alta. En contraposición, al crear una muestra uniforme entre $170$ y $180$ y evaluarla para esos mismos parámetros de verosimilitud, se esperaría que sea menos probable. Todo lo anterior debe reflejarse con ejemplos que muestren el correcto funcionamiento (10 puntos)

```python
# Se hace un muestreo normal para comprobar el ejemplo
sigma_ejemplo_normal = 1
mu_ejemplo_normal = 177

# La cantidad de observaciones es pequeña
N_ejemplo = 10
muestreo_normal_a_evaluar = muestreo_normal(N_ejemplo, mu_ejemplo_normal, sigma_ejemplo_normal)

# Se hace un muestreo uniforma para comprobar el ejemplo
minima_ejemplo = 170
maxima_ejemplo = 180
muestreo_uniforme_a_evaluar = muestreo_uniforme(N_ejemplo, minima_ejemplo, maxima_ejemplo)

mu_similitud = 176
sigma_similitud = 3

def funcion_verosimilitud_normal(muestra, mu, sigma):
    """
    Función de verosimilitud para distribución Gaussiana sin optimización logarítmica. Basada en la ecuación del material
    
    Parámetros:
    - muestra: muestra
    - mu: media de la distribución
    - sigma: desviación estándar
    
    Retorna:
    - verosimilitud: producto de las densidades gaussianas individuales
    """
    M = len(muestra)
    verosimilitud = 1.0
    
    # Calcular el producto de las densidades gaussianas
    for n in range(M):
        # Función de densidad gaussiana
        coef_normalizacion = 1 / np.sqrt(2 * np.pi * sigma**2)
        exponente = np.exp(-0.5 * ((muestra[n] - mu) / sigma)**2)
        densidad_gaussiana = coef_normalizacion * exponente
        
        verosimilitud *= densidad_gaussiana
    
    return verosimilitud

# Ejemplo de uso
print("Siguiendo los ejemplos del enunciado")
print(f"Resultado fn verosimilitud para un muestreo normal = {funcion_verosimilitud_normal(muestreo_normal_a_evaluar, mu_similitud, sigma_similitud)}")
print(f"Resultado fn verosimilitud para un muestreo uniforme = {funcion_verosimilitud_normal(muestreo_uniforme_a_evaluar, mu_similitud, sigma_similitud)}")
```

2. ¿Cual sería una forma de solucionar este problema?

**Respuesta**
Una forma de solucionar esto es aplicando el logaritmo natural en la funcion de verosimilitud (explicado en clase) evitando el problema de underflow. La siguiente celda realiza el mismo ejemplo anterior optimizando la funcion de verosimilitud de esta forma.

```python
def funcion_verosimilitud_logaritmo_natural(muestra, media, desviacion):
    M = len(muestra)
    suma_cuadrados = np.sum((muestra - media)**2)
    resultado = (-1/(2*desviacion**2)) * suma_cuadrados - (M/2) * np.log(2*np.pi) - M * np.log(desviacion)
    return resultado

print("Siguiendo los ejemplos del enunciado")
print(f"Resultado fn similitud optimizada para un muestreo normal= {funcion_verosimilitud_logaritmo_natural(muestreo_normal_a_evaluar, mu_similitud, sigma_similitud)}")
```

4. Implemente una función para determinar los valores óptimos para maximizar la función de verosimilitud normal, según teoría vista en clase. Suponga que la cantidad de observaciones (N) siempre es alta. Igualmente se espera que se provean ejemplos en el código para mostrar el correcto funcionamiento (10 puntos)

```python
def maximizar_verosimilitud_normal(muestra):
    N = len(muestra)
    
    # Estimador de máxima verosimilitud para mu
    mu_ml = np.mean(muestra)
    
    # Estimador de máxima verosimilitud para sigma^2 (sesgado)
    varianza_ml = np.sum((muestra - mu_ml)**2) / N
    sigma_ml = np.sqrt(varianza_ml)
    
    return mu_ml, sigma_ml

# Ejemplo de uso
mu = 170
sigma = 5
N = 100
muestra_ejemplo = muestreo_normal(N, mu, sigma)

mu_estimada, sigma_estimada = maximizar_verosimilitud_normal(muestra_ejemplo)

print("Ejemplo de maximización de verosimilitud:")
print(f"Muestra generada con mu={mu}, sigma={sigma}")
print("Parámetros estimados:")
print(f"mu estimada: {mu_estimada}")
print(f"sigma estimada: {sigma_estimada}")
print()
print("Comparación con parámetros reales:")
print(f"Error en mu: {mu_estimada - 170}")
print(f"Error en sigma: {sigma_estimada - 5}")
```

```python
# Parámetros para la simulación
num_muestras = 10000
N = 30
mu_real = 170
sigma_real = 1
varianza_real = sigma_real**2

# Generar 10,000 muestras y calcular la varianza estimada para cada una
varianzas_estimadas = []

for i in range(num_muestras):
    # Generar muestra de 30 observaciones
    muestra = muestreo_normal(N, mu_real, sigma_real)
    
    # Calcular estimadores de máxima verosimilitud
    mu_ml, sigma_ml = maximizar_verosimilitud_normal(muestra)
    
    # Guardar la varianza estimada (sesgada)
    varianza_ml = sigma_ml**2
    varianzas_estimadas.append(varianza_ml)

# Calcular el valor esperado de la varianza estimada
valor_esperado_varianza = np.mean(varianzas_estimadas)

# Mostrar resultados
print(f"Simulación con {num_muestras:,} muestras de {N} observaciones cada una")
print(f"Parámetros de generación: mu={mu_real}, sigma={sigma_real} (sigma^2={varianza_real})")
print()
print("Resultados del estimador sesgado de la varianza:")
print(f"Valor esperado de la varianza estimada: {valor_esperado_varianza:.4f}")
print(f"Varianza real utilizada: {varianza_real:.4f}")
print(f"Diferencia (sesgo): {valor_esperado_varianza - varianza_real:.4f}")
print()

# Análisis teórico
factor_correccion = N / (N - 1)
varianza_corregida = valor_esperado_varianza * factor_correccion

print("Análisis teórico:")
print(f"Factor de corrección esperado: N/(N-1) = {N}/{N-1} = {factor_correccion:.4f}")
print(f"Varianza corregida (no sesgada): {varianza_corregida:.4f}")
print()

print("Conclusión:")
print("El estimador sesgado subestima la varianza real.")
print("Es necesario aplicar la corrección N/(N-1) para obtener un estimador no sesgado.")
```
