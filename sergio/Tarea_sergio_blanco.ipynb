## Ejercicio 1: Análisis de Ventas de Bienes Raíces

Una agencia de bienes raíces tiene una base de datos de propiedades vendidas. Basado en los datos de la siguiente tabla, responda los apartados.

| | Menos de 30 días | De 31 a 90 | Más de 90 | Total |
|--- | --- | --- | --- |---|
|**Menos de \$50,000**|50|40|10|100|
|**50,000 a 99,999**|20|150|80|250|
|**100,000 a 149,999**|20|280|100|400|
|**Más de 150,000**|10|10|30|50|
|**Total**|100|480|220|800|

---

### 1. Estimación de la probabilidad de A (Venta en más de 90 días)

Sea $A$ el evento de vender una casa en más de 90 días.

$$ P(A) = \frac{\text{Casas vendidas en más de 90 días}}{\text{Total de casas}} $$

$$ P(A) = \frac{220}{800} = 0.275 $$

**La probabilidad de $A$ es 0.275 (27.5%).**

### 2. Estimación de la probabilidad de B (Venta en menos de $50,000)

Sea $B$ el evento de vender una casa en menos de $50,000.

$$ P(B) = \frac{\text{Casas vendidas en menos de \$50,000}}{\text{Total de casas}} $$

$$ P(B) = \frac{100}{800} = 0.125 $$

**La probabilidad de $B$ es 0.125 (12.5%).**

### 3. Probabilidad de que A y B ocurran juntos

Buscamos la probabilidad de la intersección $P(A \cap B)$. Esto corresponde a las casas que se vendieron en más de 90 días Y por menos de $50,000.

$$ P(A \cap B) = \frac{\text{Casas > 90 días y < \$50k}}{\text{Total de casas}} $$

$$ P(A \cap B) = \frac{10}{800} = 0.0125 $$

**La probabilidad conjunta es 0.0125 (1.25%).**

### 4. Probabilidad condicional: Venta en 90 días o menos dado precio < $50,000

Si definimos el evento $C$ como venderse en 90 días o menos, esto es el complemento de $A$ ($A^c$).
Se nos pide la probabilidad de $C$ dado que ocurrió $B$ (precio < $50,000).

$$ P(A^c | B) = \frac{P(A^c \cap B)}{P(B)} $$

Observando la tabla para la fila de "Menos de $50,000":
- Total en fila ($B$) = 100
- Ventas en 90 días o menos ($A^c \cap B$) = 50 (Menos de 30) + 40 (31 a 90) = 90

$$ P(A^c | B) = \frac{90}{100} = 0.9 $$

**La probabilidad es 0.9 (90%).**

### 5. Independencia de los eventos A y B

Dos eventos $A$ y $B$ son independientes si y solo si:
$$ P(A \cap B) = P(A) \cdot P(B) $$

Calculamos ambos lados:
1. $P(A \cap B) = 0.0125$ (del punto 3)
2. $P(A) \cdot P(B) = 0.275 \cdot 0.125 = 0.034375$

Comparamos:
$$ 0.0125 \neq 0.034375 $$

**Conclusión:** Dado que $P(A \cap B) \neq P(A) \cdot P(B)$, **los eventos A y B NO son independientes.**

1. Muestreo de estaturas. Los estudiantes deberán programar dos funciones para crear muestras de $N$ personas (observaciones) con una sola variable aleatoria: estatura. Para efectos prácticos, ambas funciones pueden retornar una simple lista con valores numéricos que representan la estatura en centímetros de diferentes personas. Para ambas funciones deberán mostrarse ejemplos de creación de muestras con al menos 100 observaciones y mostrarlas en un histograma. La cantidad de cubetas deberá limitarse a un máximo de 9. Las funciones a programar serán las siguientes:
 * **Muestreo Uniforme**. Además de recibir la cantidad de observaciones a generar $N$, recibirá la estatura mínima y máxima  por parámetro. Por ejemplo, si se pide una distribución uniforme entre 150 y 168, las estaturas entre ese rango deberán ser igual de probables de muestrear. (5 puntos)
  * **Muestreo Normal**. Además de recibir la cantidad de observaciones a generar $N$, recibirá la media y desviación estándar por parámetro. De esta forma, la función generará $N$ observaciones provenientes de una distribución Gaussiana con la media y desviación estándar recibidas. (5 puntos)
  

```python
import random
import matplotlib.pyplot as plt
import math

def muestreo_uniforme(N, min_height, max_height, generar_grafico=True):
    """
    Genera N observaciones de una distribucion uniforme y muestra un histograma.
    Retorna la lista de estaturas generadas.
    """
    estaturas = [random.uniform(min_height, max_height) for _ in range(N)]

    plt.figure(figsize=(8, 6))
    plt.hist(estaturas, bins=9, color="skyblue", edgecolor="black")
    plt.title(f"Muestreo Uniforme (N={N}, min={min_height}, max={max_height})")
    plt.xlabel("Estatura (cm)")
    plt.ylabel("Frecuencia")
    plt.grid(True, alpha=0.3)
    if generar_grafico:
        plt.show()
    plt.close()

    return estaturas



def muestreo_normal(N, mean, std_dev, generar_grafico=True):
    """
    Genera N observaciones de una distribucion normal y muestra un histograma.
    Retorna la lista de estaturas generadas.
    """
    estaturas = [random.gauss(mean, std_dev) for _ in range(N)]

    plt.figure(figsize=(8, 6))
    plt.hist(estaturas, bins=9, color="lightgreen", edgecolor="black")
    plt.title(f"Muestreo Normal (N={N}, media={mean}, desv={std_dev})")
    plt.xlabel("Estatura (cm)")
    plt.ylabel("Frecuencia")
    plt.grid(True, alpha=0.3)
    if generar_grafico:
        plt.show()
    plt.close()

    return estaturas

print("--- Ejemplo Muestreo Uniforme ---")
uniformes = muestreo_uniforme(100, 150, 168)
print(f"Primeras 10 estaturas uniformes: {uniformes[:10]}")
print("\n")

# Ejemplo Muestreo Normal
print("--- Ejemplo Muestreo Normal ---")
# Asumiendo una media de 170cm y desviacion estandar de 10cm
normales = muestreo_normal(100, 170, 10)
print(f"Primeras 10 estaturas normales: {normales[:10]}")
```

Deberán proveer una función que reciba una de las **muestras** del punto 1 y evalúe la probabilidad que esa muestra provenga de una distribución normal con parámetros $\mu$ y $\sigma$ (que deberán ser enviados a la función). En contraposición, al crear una muestra uniforme entre $170$ y $180$ y evaluarla para esos mismos parámetros de verosimilitud, se esperaría que sea menos probable. Todo lo anterior debe reflejarse con ejemplos que muestren el correcto funcionamiento (10 puntos)


```python


def calcular_verosimilitud(muestra, mu, sigma):
    """
    Calcula la verosimilitud (likelihood) de que una muestra provenga
    de una distribucion normal con parametros mu y sigma.

    L(mu, sigma | x) = prod( f(x_i | mu, sigma) )
    donde f es la densidad de probabilidad normal.
    """
    likelihood = 1.0
    # Usamos una variable para detectar underflow potentialmente,
    # aunque con N=100 y float64 deberia aguantar hasta ~10^-300.

    for x in muestra:
        # Funcion de densidad de probabilidad (PDF) normal
        term1 = 1 / (sigma * math.sqrt(2 * math.pi))
        term2 = math.exp(-0.5 * ((x - mu) / sigma) ** 2)
        pdf_val = term1 * term2

        likelihood *= pdf_val

    return likelihood


#Parametros para la evaluacion (Hipotesis)
mu_target = 176
sigma_target = 3
print(f"Evaluando verosimilitud para Hipotesis Normal(mu={mu_target}, sigma={sigma_target})") 

 # 1. Muestra Normal que "se parece" a la hipotesis
    # Generamos muestra con mu=177, sigma=1 (Cercano a 176, 3)
    # Nota: Usamos N=50 para evitar valores extremadamente pequeños de verosimilitud en la visualizacion,
    # aunque Python maneja floats pequeños bien.
    # 
N_samples = 50
print(f"\nGenerando muestra Normal (N={N_samples}, mu=177, sigma=1)...")
muestra_normal = muestreo_normal(N_samples, 177, 1)

    # Evaluar
L_normal = calcular_verosimilitud(muestra_normal, mu_target, sigma_target)
print(f"Verosimilitud (Likelihood) muestra Normal: {L_normal:.5e}")

    # 2. Muestra Uniforme que NO "se parece" tanto (plana entre 170 y 180)
print(f"\nGenerando muestra Uniforme (N={N_samples}, min=170, max=180)...")
muestra_uniforme = muestreo_uniforme(N_samples, 170, 180)

    # Evaluar
L_uniforme = calcular_verosimilitud(muestra_uniforme, mu_target, sigma_target)
print(f"Verosimilitud (Likelihood) muestra Uniforme: {L_uniforme:.5e}")

    # Comparacion
print("\n--- Resultados ---")
if L_normal > L_uniforme:
    print(f"La muestra NORMAL tiene MAYOR verosimilitud ({L_normal:.3e} > {L_uniforme:.3e}) de provenir de N({mu_target}, {sigma_target})."        )
    print("Esto es esperado pues la muestra normal esta mas concentrada cerca de la media hipotetica." )
else:
    print(f"La muestra UNIFORME tiene MAYOR verosimilitud ({L_uniforme:.3e} > {L_normal:.3e}).")
    print( "Esto podria pasar si la muestra uniforme cubre mejor el rango de alta probabilidad o por azar.")
```

## Respuestas a las preguntas. 

1. ¿Que problema ocurre con el calculo de la funcion de verosimilitud cuando aumentamos la cantidad de observaciones en varios ordenes de magnitud?

Ocurre un problema de desbordamiento inferior  (Underflow) numérico, esto sucede cuando el resultado de una operación aritmética es un número tan pequeño (tan cercano a cero) que la computadora no tiene suficiente precisión para representarlo.

El caso de la verosimilitud es el producto de muchas probabilidades (valores entre 0 y 1), el resultado tiende a cero muy rapidamente. Por lo que con una gran cantidad de datos, el valor se vuelve mas pequeño que el número positivo más pequeño representable por la computadora, y se redondea a 0.0, perdiendo toda la información.

2. ¿Cual seria una forma de solucionar este problema?

La solucion estandar es utilizar el "Logaritmo de la Verosimilitud" (Log-Likelihood). Al aplicar logaritmos, el producto de probabilidades se convierte en una suma de log-probabilidades. Esto transforma números muy pequeños en numeros negativos manejables (ej. $\ln(10^{-100}) = -230.2$ ), evitando el "underflow" y facilitando los cálculos numéricos y la optimización.




```python

def calcular_log_verosimilitud(muestra, mu, sigma):
    """
    Calcula el Log-Likelihood para evitar problemas numéricos (Underflow).

    l(mu, sigma | x) = sum( ln( f(x_i | mu, sigma) ) )
    """
    log_likelihood = 0.0
    # Constantes pre-calculadas para eficiencia (aunque python math lo maneja bien)
    const_term = math.log(sigma * math.sqrt(2 * math.pi))

    for x in muestra:
        # ln(pdf) = -ln(sigma * sqrt(2pi)) - 0.5 * ((x-mu)/sigma)^2
        term_exp = 0.5 * ((x - mu) / sigma) ** 2
        log_pdf_val = -const_term - term_exp
        log_likelihood += log_pdf_val

    return log_likelihood

print("--- Evaluacion de Verosimilitud ---\n")

mu_target = 176
sigma_target = 3
print(f"Evaluando verosimilitud para Hipotesis Normal(mu={mu_target}, sigma={sigma_target})")

    # -----------------------------------------------------------
    # CASO 1: N Pequeno (50) - Verosimilitud Normal funciona bien
    # -----------------------------------------------------------
N_samples = 50
print(f"#"*20)
print(f"\n[CASO 1] N={N_samples}")
print(f"#"*20)
muestra_normal = muestreo_normal(N_samples, 177, 1)
muestra_uniforme = muestreo_uniforme(N_samples, 170, 180)

L_normal = calcular_verosimilitud(muestra_normal, mu_target, sigma_target)
L_uniforme = calcular_verosimilitud(muestra_uniforme, mu_target, sigma_target)

print(f"Likelihood Normal:   {L_normal:.5e}")
print(f"Likelihood Uniforme: {L_uniforme:.5e}")

    # -----------------------------------------------------------
    # CASO 2: N Grande (2000) - Verosimilitud sufre Underflow
    # -----------------------------------------------------------
print("\n\n\n")
N_large = 2000
print(f"#"*20)
print(f"\n[CASO 2] N={N_large} (Demostracion de Underflow y Solucion)")
print(f"#"*20)
muestra_grande = muestreo_normal(N_large, 177, 1)

    # Intento con verosimilitud estandar
L_grande = calcular_verosimilitud(muestra_grande, mu_target, sigma_target)
print(f"Likelihood Estandar: {L_grande}  <-- Probable UNDERFLOW (0.0)")

    # Solucion con Log-Likelihood
LL_grande = calcular_log_verosimilitud(muestra_grande, mu_target, sigma_target)
print(f"Log-Likelihood:      {LL_grande:.4f} <-- Valor numerico valido")

    # Comparacion usando Log-Likelihood con la muestra uniforme grande
muestra_uniforme_grande = muestreo_uniforme(N_large, 170, 180)
LL_uniforme_grande = calcular_log_verosimilitud(muestra_uniforme_grande, mu_target, sigma_target)
print(f"Log-Likelihood Unif: {LL_uniforme_grande:.4f}")

if LL_grande > LL_uniforme_grande:
    print(f"Conclusion: Log-Likelihood preserva la comparacion: Normal ({LL_grande:.2f}) > Uniforme ({LL_uniforme_grande:.2f})")


```

```python

def calcular_mle_normal(muestra):
    """
    Calcula los valores optimos de mu y sigma que maximizan la funcion de
    verosimilitud normal para la muestra dada (Maximum Likelihood Estimation).

    Basado en la teoria (derivada del log-likelihood = 0):
    mu_optima = Promedio(x)
    sigma_optima = sqrt( Promedio( (x - mu_optima)^2 ) )  [Desviacion estandar poblacional/sesgada]
                   (Para N grande es equivalente a la muestral)

    Args:
        muestra (list): Lista de observaciones.

    Returns:
        tuple: (mu_optima, sigma_optima)
    """
    N = len(muestra)
    if N == 0:
        return 0.0, 0.0

    # 1. Calculo de mu (Media Muestral)
    mu_opt = sum(muestra) / N

    # 2. Calculo de sigma (Desviacion estandar poblacional para MLE)
    # Suma de cuadrados de diferencias
    suma_diff_cuadrado = sum((x - mu_opt) ** 2 for x in muestra)

    # Varianza = Suma / N (No N-1, segun definicion estricta de MLE para Normal)
    var_opt = suma_diff_cuadrado / N
    sigma_opt = math.sqrt(var_opt)

    return mu_opt, sigma_opt



print("--- Evaluacion de Verosimilitud y MLE ---\n")

mu_target = 176
sigma_target = 3

    # -----------------------------------------------------------
    # CASO 1: N Pequeno (50)
    # -----------------------------------------------------------
    # ... (Codigo anterior omitido por brevedad en display, pero mantenido en logica si se requiere) ...
    # Para consistencia con la tarea anterior, mantengo los ejemplos anteriores resumidos o los vuelvo a correr.

    # ... Re-running previous examples quickly


print(f"#"*20)
print(f"CASO 1: N Pequeno (50)")
print(f"#"*20)

N_samples = 50
muestra_normal = muestreo_normal(N_samples, 177, 1)
muestra_uniforme = muestreo_uniforme(N_samples, 170, 180)

L_normal = calcular_verosimilitud(muestra_normal, mu_target, sigma_target)
L_uniforme = calcular_verosimilitud(muestra_uniforme, mu_target, sigma_target)

    # -----------------------------------------------------------
    # CASO 2: N Grande (2000) - Log Likelihood
    # -----------------------------------------------------------

print("\n\n\n")
print(f"#"*20)
print(f"CASO 2: N Grande (2000)")
print(f"#"*20)

N_large = 2000
muestra_grande = muestreo_normal(N_large, 177, 1)
LL_grande = calcular_log_verosimilitud(muestra_grande, mu_target, sigma_target)

muestra_uniforme_grande = muestreo_uniforme(N_large, 170, 180)
LL_uniforme_grande = calcular_log_verosimilitud(muestra_uniforme_grande, mu_target, sigma_target)

print(f"Log-Likelihood Comparacion (N={N_large}): Normal={LL_grande:.2f} vs Uniforme={LL_uniforme_grande:.2f}")

    # -----------------------------------------------------------
    # CASO 3: MLE (Maximum Likelihood Estimation)
    # -----------------------------------------------------------

print("\n\n\n")
print(f"#"*20)
print(f"CASO 3: MLE (Maximum Likelihood Estimation)")
print(f"#"*20)

    # Generamos datos con parametros CONOCIDOS
mu_real = 170.0
sigma_real = 5.0
print(f"Generando muestra de N={N_large} con parametros REALES: mu={mu_real}, sigma={sigma_real}")

muestra_mle = muestreo_normal(N_large, mu_real, sigma_real)

    # Calculamos los parametros optimos recuperados
mu_estimada, sigma_estimada = calcular_mle_normal(muestra_mle)

print("\n\n\n")
print(f"#"*20)
print(f"Parametros estimados por MLE:")
print(f"#"*20)


print(f"  mu_estimada:    {mu_estimada:.4f}  (Error: {abs(mu_estimada - mu_real):.4f})")
print(f"  sigma_estimada: {sigma_estimada:.4f}  (Error: {abs(sigma_estimada - sigma_real):.4f})")

print("\nVerificacion:")
if abs(mu_estimada - mu_real) < 0.5 and abs(sigma_estimada - sigma_real) < 0.5:
    print("EXITO: Los parametros estimados estan muy cerca de los reales.")
else:
    print("ALERTA: La estimacion difiere mas de lo esperado.")
```

La corrección consiste en multiplicar por el factor de Bessel: N / (N-1). Esto transforma la fórmula a: S_unbiased^2 = 1/(N-1) * sum(x-mu)^2. En este caso: 0.966 * (30/29) = 1.0.


```python
print("--- Demostracion de Sesgo en Estimador de Varianza (Refactorizado) ---\n")

# Parametros de Generacion
N_muestras = 10000
N_obs = 30
mu_real = 170
sigma_real = 1
varianza_real = sigma_real**2

print(f"Generando {N_muestras} muestras de {N_obs} observaciones cada una...")
print(f"Distribucion Base: Normal(mu={mu_real}, sigma={sigma_real}) -> Varianza Real = {varianza_real}\n")

# Lista para acumular las varianzas estimadas (sesgadas)
lista_varianzas_sesgadas = []

for c in range(N_muestras):
# 1. Generar Muestra usando programa1 (Sin Plot!)
   muestra = muestreo_normal(N_obs, mu_real, sigma_real, generar_grafico=False)

  # 2. Calcular MLE usando programa2
# calcular_mle_normal devuelve (mu_opt, sigma_opt)
   c, sigma_opt = calcular_mle_normal(muestra)

# La varianza sesgada es sigma_opt^2
   varianza_sesgada = sigma_opt**2

   lista_varianzas_sesgadas.append(varianza_sesgada)

# 3. Calcular Valor Esperado (Promedio de las varianzas)
valor_esperado_estimador = sum(lista_varianzas_sesgadas) / N_muestras

print(f"Resultados:")
print(f"--------------------------------------------------")
print(f"Varianza Real (Target):  {varianza_real:.4f}")
print(f"Promedio de Varianzas Estimadas: {valor_esperado_estimador:.4f}")

diferencia = valor_esperado_estimador - varianza_real
print(f"Diferencia (Sesgo):  {diferencia:.4f}")

teorico_esperado = (N_obs - 1) / N_obs * varianza_real
print(f"Valor Teorico Esperado ((N-1)/N):{teorico_esperado:.4f}")

print(f"--------------------------------------------------")

if abs(valor_esperado_estimador - teorico_esperado) < 0.01:
   print("CONCLUSION: La simulacion confirma la teoria. El estimador esta sesgado.")
   print(f"Se obtuvo ~{valor_esperado_estimador:.3f} en lugar de {varianza_real}.")
   print(f"Factor de correccion necesario: N/(N-1) = {N_obs}/{N_obs - 1} = {N_obs / (N_obs - 1):.4f}")
   correccion = valor_esperado_estimador * (N_obs / (N_obs - 1))
   print(f"Varianza corregida: {correccion:.4f}")
else:
   print("ALERTA: Resultado inseperado.")

```
